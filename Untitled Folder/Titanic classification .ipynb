# Prepare the data and feature engineering

## feature combination

```python
#Family size
dataset['c_Family']=dataset['Parch']+dataset['SibSp']+1
dataset
```

## Handling missing values 

```python
klib.corr_plot(dataset,target="Age")
```

```python
# Age and Cabin are the only features with meassing values. 
# cabin same values as classe donc on ne la garde pas 
# Preprocessing de Age

## Fill Age with the median age of similar rows according to Pclass, Parch and SibSp
# Index of NaN age rows
index_NaN_age = list(dataset["Age"][dataset["Age"].isnull()].index)

for i in index_NaN_age :
    age_med = dataset["Age"].median()
    age_pred = dataset["Age"][(dataset['c_Family'] == dataset.iloc[i]["c_Family"])].median()
    if not np.isnan(age_pred) :
        dataset['Age'].iloc[i] = age_pred
    else :
        dataset['Age'].iloc[i] = age_med
```

```python
klib.missingval_plot(dataset["Age"])
```

## Handling test and categorical attributes

```python
dataset.info()
```

```python
dataset.drop(labels = ["PassengerId"], axis = 1, inplace = True)
dataset.drop(labels = ["Name"], axis = 1, inplace = True)
dataset.drop(labels = ["Ticket"], axis = 1, inplace = True)
dataset.drop(labels = ["Cabin"], axis = 1, inplace = True)
```

```python
# Create categorical values for Pclass
dataset["Pclass"] = dataset["Pclass"].astype("category")
dataset = pd.get_dummies(dataset, columns = ["Pclass"],prefix="Pc")
dataset["Embarked"] = dataset["Embarked"].astype("category")
dataset = pd.get_dummies(dataset, columns = ["Embarked"],prefix="Em")
```

```python
dataset["Sex"] = dataset["Sex"].astype("category")
dataset = pd.get_dummies(dataset, columns = ["Sex"],prefix="Sex")
```

```python
klib.missingval_plot(dataset)
```

```python
dataset.info()
```

## Feature scaling

```python
scaler = preprocessing.StandardScaler().fit(dataset[["Age","SibSp","Parch","Fare"]])
dataset[["Age","SibSp","Parch","Fare"]] = scaler.transform(dataset[["Age","SibSp","Parch","Fare"]])
```

```python
train, test = train_test_split(dataset, test_size=0.1, random_state=42)
```

```python
X_train=train.drop(labels = ["Survived"], axis = 1)
Y_train=train['Survived'].copy()
X_train
```
